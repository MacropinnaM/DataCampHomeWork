## Dictionaries

### Creating dictionary

```python
# Lists

countries = ['Russia', 'Japan', 'Croatia', 'Poland']
languages = ['Russian', 'Japan', 'Croatian', 'Polish']
currencies = ['Ruble', 'Yen', 'Kuna', 'Zloty']

ind_rus = countries.index('Russia')
print('Russia: ', languages[ind_rus], currencies[ind_rus])

ind_jap = countries.index('Japan')
print('Japan: ', languages[ind_jap], currencies[ind_jap])

ind_cro = countries.index('Croatia')
print('Croatia: ', languages[ind_cro], currencies[ind_cro])

ind_pol = countries.index('Poland')
print('Poland: ', languages[ind_pol], currencies[ind_pol])
```

```python
# Lists to dictionary

geo_dict = {
    countries[0]:[languages[0], currencies[0]],
    countries[1]:[languages[1], currencies[1]],
    countries[2]:[languages[2], currencies[2]],
    countries[3]:[languages[3], currencies[3]],
}

print(geo_dict)
```

```python
# Access to dictionary

geo_dict = {'Russia': ['Russian', 'Ruble'], 
            'Japan': ['Japan', 'Yen'], 
            'Croatia': ['Croatian', 'Kuna'], 
            'Poland': ['Polish', 'Zloty']}

print(geo_dict.keys())
print(geo_dict.values())

print(geo_dict['Croatia'])
print(geo_dict['Poland'][0], geo_dict['Poland'][1])
```

### Dictionary manipulation

```python
# Add new items

geo_dict = {'Russia': ['Russian', 'Ruble'], 
            'Japan': ['Japan', 'Yen'], 
            'Croatia': ['Croatian', 'Kuna'], 
            'Poland': ['Polish', 'Zloty']}

geo_dict['Sweden'] = ['Swedish', 'Krona']
print('Sweden' in geo_dict)
print('Sweden: ', geo_dict['Sweden'])
```

```python
# Remove item

geo_dict = {'Croatia': ['Croatian', 'Kuna'],
             'Japan': ['Japan', 'Yen'],
             'Poland': ['Polish', 'Zloty'],
             'Russia': ['Russian', 'Ruble'],
             'Sweden': ['Swedish', 'Krona']}

del(geo_dict['Russia'])
print('Russia' in geo_dict)
```

```python
# Update items

geo_dict = {'Croatia': ['Croatian', 'Kuna'],
             'Japan': ['Japan', 'Yen'],
             'Poland': ['Polish', 'Zloty'],
             'Russia': ['Russian', 'Ruble'],
             'Sweden': ['Swedish', 'Krona']}

geo_dict['Croatia'] = ['Croatian', 'Euro']

print(geo_dict)
```

## Logic in Python

### Comparision operators

```python
# Equal and not equal

print(True == False)
print(26 * 8 != 4)
print('Cat' == 'kitten')
print(True == 1)
print('Dog' != True)
```

```python
# Greater than and less than

print(26 > 4)
print(8 <= 7 + 2)

y = 'meow'
print('meow' >= y)
```

```python
# Compare arrays

heights = np.array([168, 172, 170, 174, 167, 164])
weights = np.array([60, 56, 62, 58, 52, 52])

print(heights > weights)
print(weights >= (heights - 110))
print(heights >= 170)
print(weights < 56)
```

### Boolean Operators

```python
# And, or, not

my_height = 166
your_height = 182

print(my_height < 170 and your_height > 170)
print(my_height < 170 or your_height < 170)
print(not your_height < 170)
```

```python
# Boolean with Numpy

heights = np.array([168, 172, 170, 174, 167, 164])
weights = np.array([60, 56, 62, 58, 52, 52])

print(np.logical_and(heights > 170, weights < 60))
print(np.logical_or(heights > 165, weights < 50))
print(np.logical_not(heights < 170))
```

### if-elif-else

```python
# if

city = 'Moscow'
pop = 12.4

if city == 'Moscow':
    print("You're in the capital of Russia")

if pop > 10.0:
    print(str(city) + " is a megacity")
```

```python
# else

city = 'Krasnodar'
pop = 0.9

if city == 'Moscow':
    print("You're in the capital of Russia")
else:
    print("You're somewhere in Russia")

if pop > 10.0:
    print(str(city) + " is a megacity")
else:
    print(str(city) + " is just a town")
```

```python
# else

city = 'Saint Petersburg'
pop = 5.3

if city == 'Moscow':
    print("You're in the capital of Russia")
elif city == 'Saint Petersburg':
    print("You're in the cultural capital of Russia")
else:
    print("You're somewhere in Russia")

if pop > 10.0:
    print(str(city) + " is a megacity")
elif pop < 10.0 and pop > 1.0:
    print(str(city) + " is a megapolis")
else:
    print(str(city) + " is just a town")
```

### Filtering pandas DataFrame

```python
# Filtering Schengen countries

df = pd.read_csv('EU_data.csv')
df.index = codes

sch = df['Schengen']
sel = df[sch]
print(sel)
```

```python
# Filtering Schengen counties with population bellow 1M

df = pd.read_csv('EU_data.csv')
df.index = codes

print("Small Schengen Countries")
sel = df[df['Schengen']]
small_schengen = sel[sel['Population'] < 1000000]
print(small_schengen)
```

```python
# Filtering Schengen counties with population over 10M

df = pd.read_csv('EU_data.csv')
df.index = codes

print("Large Schengen Countries")
sel = df[df['Schengen']]
large_schengen = sel[sel['Population'] > 1000000]
print(large_schengen)
```

```python
# Filtering Schengen counties with population between 1M and 10M

df = pd.read_csv('EU_data.csv')
df.index = codes

print("Medium Schengen Countries")
sel = df[df['Schengen']]
between = np.logical_and(sel['Population'] > 1000000, sel['Population'] < 10000000)
medium_schengen = sel[between]
print(medium_schengen)
```

## Loops

### while loop

```python
# Basic while loop

x = 4
while x > 0:
    print(x)
    print('step to zero...')
    x = x - 1
```

```python
# While loop with conditions

x = -4
while x != 0:
    if x > 0:
        print(x)
        print('step to zero...')
        x = x - 1
    else:
        print(x)
        print('step to zero...')
        x = x + 1
```

### for loop

```python
# Basic for loop

areas = [82409,  30280, 108560,  55960,   9251,  77240]

for area in areas:
    print(area)
```

```python
# Indexes and values

areas = [82409,  30280, 108560,  55960,   9251,  77240]

for ind, area in enumerate(areas):
    print('area ' + str(ind + 1) + ': ' + str(area))
```

```python
# Make list of lists

areas = [82409,  30280, 108560,  55960,   9251,  77240]
countries = ['Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czech Republic']

countries_areas = []
for ind, country in enumerate(countries):
    new_list = [country, areas[ind]]
    countries_areas.append(new_list)
print(countries_areas)
```

```python
# Loop over list of lists

c_areas = [['Austria', 82409], ['Belgium', 30280], ['Bulgaria', 108560], 
           ['Croatia', 55960], ['Cyprus', 9251], ['Czech Republic', 77240]]

for x in c_areas:
    print('Area of ' + x[0] + ': ' + str(x[1]))
```

### Loop over data structures

```python
# Loop over dictionary

c_areas = {'Austria': 82409, 'Belgium': 30280, 'Bulgaria': 108560, 
           'Croatia': 55960, 'Cyprus': 9251, 'Czech Republic': 77240}
print(c_areas.items())
print()

for x, y in c_areas.items() :
    print('Area of ' + x + ': ' + str(y))
```

```python
# Loop over np.array

np_areas = np.array(areas)

for x in np.nditer(np_areas):
    print(x)
```
